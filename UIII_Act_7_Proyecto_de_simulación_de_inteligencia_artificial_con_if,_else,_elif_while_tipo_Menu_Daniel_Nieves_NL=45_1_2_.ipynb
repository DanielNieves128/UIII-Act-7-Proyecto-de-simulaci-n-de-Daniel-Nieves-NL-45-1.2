```python
print("======================================")
print("   ENTRENADOR IA - MODO DEPORTISTA   ")
print("======================================")

nombre = input("¿Cuál es tu nombre?: ")
deporte = input("¿Qué deporte practicas?: ")

print("\nHola", nombre)
print("Soy tu entrenador inteligente de", deporte)

opcion = ""

while opcion != "5":

    print("\n========== MENÚ PRINCIPAL ==========")
    print("1. Evaluar energía")
    print("2. Plan de entrenamiento")
    print("3. Consejos de alimentación")
    print("4. Mensaje motivador")
    print("5. Salir")

    opcion = input("Elige una opción (1-5): ")

    # OPCIÓN 1
    if opcion == "1":

        energia = input("¿Cómo te sientes hoy? (alta/media/baja): ")

        if energia == "alta":
            print("IA: Excelente. Hoy puedes entrenar fuerte.")

        elif energia == "media":
            print("IA: Buen trabajo. Entrena con equilibrio.")

        elif energia == "baja":
            print("IA: Descansa un poco y toma agua.")

        else:
            print("IA: No entendí tu respuesta.")

    # OPCIÓN 2
    elif opcion == "2":

        objetivo = input("¿Qué quieres mejorar? (fuerza/velocidad/resistencia): ")

        if objetivo == "fuerza":
            print("IA: Haz lagartijas, sentadillas y abdominales.")

        elif objetivo == "velocidad":
            print("IA: Practica sprints cortos y saltos.")

        elif objetivo == "resistencia":
            print("IA: Corre 20 minutos sin detenerte.")

        else:
            print("IA: Objetivo no reconocido.")

    # OPCIÓN 3
    elif opcion == "3":

        comida = input("¿Comes saludable? (si/no): ")

        if comida == "si":
            print("IA: Muy bien. Sigue comiendo frutas y verduras.")

        else:
            print("IA: Intenta evitar comida chatarra.")

    # OPCIÓN 4
    elif opcion == "4":

        animo = input("¿Necesitas motivación? (si/no): ")

        if animo == "si":
            print("IA: ¡Nunca te rindas! Tú puedes lograrlo.")

        else:
            print("IA: Sigue trabajando duro cada día.")

    # OPCIÓN 5
    elif opcion == "5":

        print("IA: Gracias por usar el Entrenador IA.")
        print("Hasta luego", nombre)

    # OPCIÓN INVÁLIDA
    else:

        print("IA: Opción inválida. Intenta otra vez.")

print("\nPrograma finalizado.")
print("Programa creado por Daniel Nieves NL = 45")
```
